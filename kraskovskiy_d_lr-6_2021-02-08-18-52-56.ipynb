## Задание 2
Исследовать функцию с помощью производной, найти необходимые пределы и решить уравнения.

Построить график функции и асимптот, отметить и подписать точки экстремума и точки перегиба,
включить функцию и асимптоты в легенду. 
$$
y=xe^{2/x} + 1 
$$

```python
from sympy.abc import x
def f(x):
    return x*sympy.E**(2/x) + 1

f_1 = f(x).diff(x, 1)
f_2 = f(x).diff(x, 2)
display(f_1, f_2)
f_1_roots = solve(f_1, x)
f_2_roots = solve(f_2, x)

X1 = np.linspace(float(f_1_roots[0] - 9), float(f_1_roots[-1] + 5), 100)
ax = plt.gca()
ax.plot(X1, f(X1), label=r'f(x) = $𝑥𝑒^{2/𝑥} + 1$')

ax.axvline(x = 0, color='red', linestyle='--', label=r'x = 0')
ax.set_ylim(-20, 20)
points_y1 = [float(f(point)) for point in f_1_roots]
points_y2 = [float(f(point)) for point in f_2_roots]
ax.scatter(f_1_roots, points_y1, 50, color='green')
ax.scatter(f_2_roots, points_y2, 50, color='red')

kpos = sympy.limit(f(x)/x, x, +sympy.oo)
kneg = sympy.limit(f(x)/x, x, -sympy.oo)
display(kpos, kneg, "Они совпадают")
b = sympy.limit(f(x) - kpos*x, x, sympy.oo)
display(b)
def y_as(x):
    return x + 3

ss = np.linspace(float(f_1_roots[0] - 9), float(f_1_roots[-1] + 5), 100)


ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
annotations1 = [r'$Extr_1$', r'$Extr_1$']
annotations2 = [r'$A_1$', r'$A_2$']

for i, x in enumerate(f_1_roots):
    ax.annotate(annotations1[i], xy=(x+0.1, points_y1[i]-0.3), xycoords='data', xytext=(-5, 10), textcoords='offset points', fontsize=16)
for i, x in enumerate(f_2_roots):
    ax.annotate(annotations2[i], xy=(x+0.1, points_y2[i]-0.3), xycoords='data', xytext=(-5, 10), textcoords='offset points', fontsize=16)
ax.plot(ss, y_as(ss), color='red', linestyle='--', label=r'y = $x + 3$')
ax.legend()

```
